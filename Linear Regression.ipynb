```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
data = pd.read_csv('C:/Users/Chinmay Gupta/Desktop/Data set/Linear Regression/train.csv')
data.head()
```

```python
data = pd.get_dummies(data)
```

```python
data.shape
```

```python
# Making the train data set
train = data.iloc[0:7999]
train
```

```python
# Making the test data set
test = data[8000:]
test
```

```python
# For using the scikit learn algo, dropping the target variable from the column we use axis=1 & making this only independent one
x_train = train.drop('Item_Outlet_Sales',axis=1)
#x_train
```

```python
# For using the scikit learn algo keeping the target variable at this place only
y_train = train[['Item_Outlet_Sales']]
#y_train
```

```python
# The test data should not contain dependent variable, so dropping the dependent variable
x_test = test.drop('Item_Outlet_Sales',axis=1)
#x_test
```

```python
# Keeping separately the target variable of the test dataset so to check it at last 
true_p = test['Item_Outlet_Sales']
```

```python
from sklearn.linear_model import LinearRegression
# Creating an object lreg 
lreg = LinearRegression()
```

```python
#x_train = pd.get_dummies(x_train)
```

```python
x_train.shape
```

```python
#x_test = pd.get_dummies(x_test)
```

```python
x_train.fillna(0,inplace=True)
```

```python
x_test.fillna(0,inplace=True)
```

```python
# Linear Regression has a function fit which fits the training data to the model
lreg.fit(x_train,y_train)
```

```python
pred= lreg.predict(x_test) 
pred
```

```python
# For checking the performance we use R-Squared method on test data
lreg.score(x_test,true_p)
```

```python
# For checking the performance we use R-Squared method on train data
lreg.score(x_train,y_train)
```

```python
# Evaluation of model - RMSE i.e (Target Actual- Target Predict). There is no use of independent variable
rmse_test = np.sqrt(np.mean(np.power((np.array(true_p)-np.array(pred)),2)))
rmse_test
```

```python
rmse_train = np.sqrt(np.mean(np.power((np.array(y_train)-np.array(lreg.predict(x_train))),2)))
rmse_train
```
